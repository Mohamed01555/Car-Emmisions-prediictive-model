```python
X = new_dataset.drop('CO2 g/km',axis = 1)
X
```

# Preprocessing

## 1- Filling NaN
## 2- Normalization
## 3- Categorical variables(Label Encoder , OneHotEncoder)

```python
X.info()
```

```python
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = 'NaN', strategy = 'mean',axis = 0 )
X[['Engine Capacity', 'Metric Urban (Cold)', 'Metric Extra-Urban', 'Metric Combined'
                         , 'Imperial Urban (Cold)', 'Imperial Extra-Urban', 'Imperial Combined', 'Emissions CO [mg/km]'
                         , 'THC Emissions [mg/km]', 'Emissions NOx [mg/km]', 'THC + NOx Emissions [mg/km]'
                         , 'Particulates [No.] [mg/km]']] = imputer.fit_transform(X[['Engine Capacity', 'Metric Urban (Cold)', 'Metric Extra-Urban', 'Metric Combined'
                         , 'Imperial Urban (Cold)', 'Imperial Extra-Urban', 'Imperial Combined', 'Emissions CO [mg/km]'
                         , 'THC Emissions [mg/km]', 'Emissions NOx [mg/km]', 'THC + NOx Emissions [mg/km]'
                         , 'Particulates [No.] [mg/km]']])
```

```python
X.info()
```

# Another easier way to fill NaN

```python
X.isna().sum()
```

```python
X = X.fillna(X.mean())
```

```python
X.info()
```

# 2- Handling categorical data

```python
from sklearn.preprocessing import LabelEncoder
lable = LabelEncoder()
X['Fuel Type'] = lable.fit_transform(X['Fuel Type'])
X
```

```python
from sklearn.preprocessing import OneHotEncoder
one = OneHotEncoder(categorical_features=[1])
new_X = one.fit_transform(X).toarray()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(new_X, Y, test_size = 0.2, random_state = 0)
```

# let's build our model


```python
from sklearn.linear_model import LinearRegression
lin = LinearRegression()

from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 1)

x_poly_train = poly_reg.fit_transform(X_train)
x_poly_test  = poly_reg.fit_transform(X_test)

lin.fit(x_poly_train, Y_train)
print(lin.coef_)
# Test model
y_pred = lin.predict(x_poly_test)

# Calculate the Accuracy
from sklearn.metrics import explained_variance_score
print('Polynomial Linear Regression Accuracy:',explained_variance_score(Y_test,y_pred))
```

# As we have observed the accuracy of linear models is 0.9959

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
print(X_train.shape,Y_train.shape)
model.coef_
```

```python
prediction = model.predict(X_test)
print(X_test.shape, prediction.shape)
from sklearn.metrics import r2_score
r2_score(Y_test,prediction)
```
